```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.style.use("ggplot")

from wordcloud import WordCloud, STOPWORDS

import seaborn as sns
```

```python
df = pd.read_csv('data/clf_data_cleaned.csv')
```

```python
fig, ax = plt.subplots()
df.groupby('Outcome Type')['Age'].mean().plot(ax=ax, kind='bar', xlim=(0, 15), figsize=(8,4), title="Average Age per Outcome Type")
ax.set_xticklabels(['Euthanize', 'Transfer', 'Return', 'Adoption'], rotation=45)
ax.set_ylabel("Average Age")
plt.legend()
```

```python
fig, ax = plt.subplots()
df.groupby('Outcome Type')['Age'].plot(ax=ax, kind='kde', xlim=(0, 10), figsize=(8,6), title="Age Distribution per Outcome Type")
ax.legend(('Euthanize', 'Transfer', 'Return', 'Adoption'))
ax.set_xlabel("Age")
```

```python
df1 = pd.read_csv('data/Animal_Shelter_Intake_and_Outcome.csv')
```

```python
df1
```

```python
df_cats = df1[df1['Type'] == 'CAT']
df_dogs = df1[df1['Type'] == 'DOG']
df_other = df1[df1['Type'] == 'OTHER']
```

```python
wordcloud_dogbreed = WordCloud(background_color='white').generate(' '.join(df_dogs['Breed']))

fig, ax = plt.subplots(figsize=(8,12))
plt.imshow(wordcloud_dogbreed)
plt.axis("off")
plt.show()
```

```python
wordcloud_catbreed = WordCloud(background_color='white').generate(' '.join(df_cats['Breed']))

fig, ax = plt.subplots(figsize=(8,12))
plt.imshow(wordcloud_catbreed)
plt.axis("off")
plt.show()
```

```python
wordcloud_otherbreed = WordCloud(background_color='white').generate(' '.join(df_other['Breed']))

fig, ax = plt.subplots(figsize=(8,12))
plt.imshow(wordcloud_otherbreed)
plt.axis("off")
plt.show()
```
